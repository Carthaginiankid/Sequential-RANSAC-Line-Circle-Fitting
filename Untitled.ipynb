```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. RANSAC Configuration and Constraints ---
T_LINE = 0.05       # 5 cm tolerance for line inliers
T_CIRCLE = 0.05     # 5 cm tolerance for circle inliers
MAX_RANSAC_ITERATIONS = 2000

MIN_INLIERS_LINE = 10
MIN_INLIERS_CIRCLE = 8
MIN_LENGTH_LINE = 0.4 
R_MIN_CIRCLE = 0.2
R_MAX_CIRCLE = 0.5
MIN_CIRCLE_ARC_DEG = 30  # Minimum arc coverage in degrees


# --- 2. Model Fitting Functions ---
def fit_line_svd(points_2d):
    if len(points_2d) < 2:
        return None, None, None
    centroid = np.mean(points_2d, axis=0)
    centered_points = points_2d - centroid
    cov = centered_points.T @ centered_points
    eigvals, eigvecs = np.linalg.eigh(cov)
    l_vec = eigvecs[:, 0]  # normal vector
    c = -np.dot(l_vec, centroid)
    return l_vec, c, centroid

def get_line_segment_length(points_2d, l_vec, centroid):
    tangent = np.array([-l_vec[1], l_vec[0]])
    tangent /= np.linalg.norm(tangent)
    projections = np.dot(points_2d - centroid, tangent)
    return np.max(projections) - np.min(projections)

def fit_circle_lls(points_2d):
    N = len(points_2d)
    if N < 3:
        return None, None
    X, Y = points_2d[:,0], points_2d[:,1]
    A = np.column_stack([np.ones(N), -2*X, -2*Y])
    B = -(X**2 + Y**2).reshape(N,1)
    try:
        P, _, _, _ = np.linalg.lstsq(A, B, rcond=None)
    except np.linalg.LinAlgError:
        return None, None
    C, x0, y0 = P.flatten()
    R_sq = x0**2 + y0**2 - C
    if R_sq <= 0: return None, None
    R = np.sqrt(R_sq)
    return np.array([x0, y0]), R

def distance_point_to_line(point, l_vec, c):
    return np.abs(np.dot(l_vec, point) + c) / np.linalg.norm(l_vec)

def distance_point_to_circle(point, center, radius):
    return np.abs(np.linalg.norm(point[:2] - center) - radius)


# --- 3. RANSAC Core Functions ---
def run_ransac_line(points_2d):
    best_inliers = []
    for _ in range(MAX_RANSAC_ITERATIONS):
        if len(points_2d) < 5: break
        sample_idx = np.random.choice(len(points_2d), 5, replace=False)
        sample = points_2d[sample_idx]
        l_vec, c, centroid = fit_line_svd(sample)
        if l_vec is None: continue
        inliers = [i for i, p in enumerate(points_2d) if distance_point_to_line(p, l_vec, c) < T_LINE]
        if len(inliers) > len(best_inliers):
            best_inliers = inliers
    if len(best_inliers) < MIN_INLIERS_LINE:
        return None, None
    inlier_pts = points_2d[best_inliers]
    l_vec, c, centroid = fit_line_svd(inlier_pts)
    length = get_line_segment_length(inlier_pts, l_vec, centroid)
    if length >= MIN_LENGTH_LINE:
        c_refined = -np.dot(l_vec, centroid)
        return {'l_vec': l_vec, 'c': c_refined, 'length': length}, best_inliers
    return None, None

def run_ransac_circle(points_2d):
    best_inliers = []
    for _ in range(MAX_RANSAC_ITERATIONS):
        if len(points_2d) < 3: break
        idx = np.random.choice(len(points_2d), 3, replace=False)
        p1, p2, p3 = points_2d[idx]
        # Check collinearity
        area = 0.5*np.abs(np.cross(p2 - p1, p3 - p1))
        if area < 1e-6: continue
        # Circle from 3 points
        A = np.array([[p2[0]-p1[0], p2[1]-p1[1]],
                      [p3[0]-p1[0], p3[1]-p1[1]]])
        B = np.array([(p2[0]**2+p2[1]**2 - p1[0]**2 - p1[1]**2)/2,
                      (p3[0]**2+p3[1]**2 - p1[0]**2 - p1[1]**2)/2])
        try:
            center = np.linalg.solve(A, B)
        except np.linalg.LinAlgError:
            continue
        R = np.linalg.norm(center - p1)
        inliers = [i for i,p in enumerate(points_2d) if distance_point_to_circle(p, center, R) < T_CIRCLE]
        if len(inliers) > len(best_inliers):
            best_inliers = inliers
    if len(best_inliers) < MIN_INLIERS_CIRCLE:
        return None, None
    inlier_pts = points_2d[best_inliers]
    center, R = fit_circle_lls(inlier_pts)
    if center is None or not (R_MIN_CIRCLE <= R <= R_MAX_CIRCLE):
        return None, None
    # Minimal arc check
    angles = np.arctan2(inlier_pts[:,1]-center[1], inlier_pts[:,0]-center[0])
    arc_deg = np.degrees(np.max(angles)-np.min(angles))
    if arc_deg < MIN_CIRCLE_ARC_DEG:
        return None, None
    return {'center': center, 'R': R}, best_inliers

def sequential_ransac(points_2d):
    remaining_idx = np.arange(len(points_2d))
    found_models = []
    while len(remaining_idx) >= max(MIN_INLIERS_LINE, MIN_INLIERS_CIRCLE):
        current_pts = points_2d[remaining_idx]
        line_model, line_inliers = run_ransac_line(current_pts)
        circle_model, circle_inliers = run_ransac_circle(current_pts)
        selected_model = None
        if line_model and circle_model:
            if len(line_inliers) >= len(circle_inliers):
                selected_model = {'type':'line', **line_model, 'inliers_idx':line_inliers}
            else:
                selected_model = {'type':'circle', **circle_model, 'inliers_idx':circle_inliers}
        elif line_model:
            selected_model = {'type':'line', **line_model, 'inliers_idx':line_inliers}
        elif circle_model:
            selected_model = {'type':'circle', **circle_model, 'inliers_idx':circle_inliers}
        if selected_model:
            original_idx = remaining_idx[selected_model['inliers_idx']]
            selected_model['original_indices'] = original_idx
            found_models.append(selected_model)
            remaining_idx = np.delete(remaining_idx, selected_model['inliers_idx'])
            print(f"Found {selected_model['type']} with {len(original_idx)} inliers. Remaining: {len(remaining_idx)}")
        else:
            break
    return found_models, remaining_idx

# --- 4. PLY Output ---
def save_to_ply(filepath, points_2d, models, remaining_idx):
    N = len(points_2d)
    colors = np.full((N,3), 255, dtype=np.uint8)
    assigned_mask = np.zeros(N, dtype=bool)
    for model in models:
        idx = model['original_indices']
        unassigned = idx[~assigned_mask[idx]]
        if model['type']=='line':
            colors[unassigned] = [255,0,0]  # Red
        else:
            colors[unassigned] = [0,0,255]  # Blue
        assigned_mask[unassigned] = True
    points_3d = np.column_stack((points_2d, np.zeros(N)))
    with open(filepath,'w') as f:
        f.write("ply\nformat ascii 1.0\n")
        f.write(f"element vertex {N}\nproperty float x\nproperty float y\nproperty float z\n")
        f.write("property uchar red\nproperty uchar green\nproperty uchar blue\nend_header\n")
        for i in range(N):
            x,y,z = points_3d[i]
            r,g,b = colors[i]
            f.write(f"{x:.6f} {y:.6f} {z:.6f} {r} {g} {b}\n")
    print(f"PLY saved to {filepath}.")


# --- 5. Run Sequential RANSAC ---
if 'points' in locals() and len(points)>0:
    OUT_FILE = 'ransac_segments.ply'
    FOUND_MODELS, FINAL_OUTLIERS = sequential_ransac(points)
    save_to_ply(OUT_FILE, points, FOUND_MODELS, FINAL_OUTLIERS)
else:
    print("Error: 'points' variable not found or empty. Load your 2D Cartesian data first.")
```

```python
# Distance between a point and a line (ax + by + c = 0)
def point_line_distance(a, b, c, points):
    x, y = points[:,0], points[:,1]
    return np.abs(a*x + b*y + c) / np.sqrt(a**2 + b**2)
# Distance between a point and a circle (center, radius)
def point_circle_distance(center, radius, points):
    return np.sqrt((points[:,0]-center[0])**2 + (points[:,1]-center[1])**2) - radius
```

```python
def fit_line_optimal(points):
    """
    Fit a line using centroid + eigenvector (total least squares)
    Returns: a, b, c for line ax + by + c = 0
    """
    centroid = points.mean(axis=0)
    centered = points - centroid
    cov = centered.T @ centered
    vals, vecs = eig(cov)
    direction = vecs[:, np.argmax(vals)]  # dominant eigenvector
    normal = np.array([-direction[1], direction[0]])  # perpendicular to line
    a, b = normal
    c = -np.dot(normal, centroid)
    return a, b, c
```

```python

def fit_circle(points):
    """
    Fit a circle to points using pseudo-inverse
    Returns: center (x0, y0), radius R
    """
    N = points.shape[0]
    A = np.column_stack((np.ones(N), -2*points[:,0], -2*points[:,1]))
    b = -(points[:,0]**2 + points[:,1]**2)
    
    x = pinv(A) @ b
    c, x0, y0 = x
    R = np.sqrt(x0**2 + y0**2 - c)
    
    return np.array([x0, y0]), R
```

```python
def sequential_ransac(points,
                      min_line_pts=10, min_line_len=0.4,
                      min_circle_pts=8, circle_r_range=(0.2,0.5),
                      distance_thresh_line=0.05,
                      distance_thresh_circle=0.02,
                      max_iterations=2000):
    
    remaining_points = points.copy()
    lines = []
    circles = []
    
    iteration = 0
    while remaining_points.shape[0] >= min(min_line_pts, min_circle_pts) and iteration < max_iterations:
        iteration += 1
        new_segment_found = False

        # ----------------- LINE FIT -----------------
        if remaining_points.shape[0] >= min_line_pts:
            a, b, c = fit_line_optimal(remaining_points)
            distances_line = point_line_distance(a, b, c, remaining_points)
            line_inliers = remaining_points[distances_line <= distance_thresh_line]

            if line_inliers.shape[0] >= min_line_pts:
                x_vals, y_vals = line_inliers[:,0], line_inliers[:,1]
                length = np.sqrt((x_vals.max()-x_vals.min())**2 + (y_vals.max()-y_vals.min())**2)
                if length >= min_line_len:
                    # Re-estimate line from all inliers (optimal)
                    a, b, c = fit_line_optimal(line_inliers)
                    lines.append((a, b, c, line_inliers))
                    remaining_points = remaining_points[distances_line > distance_thresh_line]
                    new_segment_found = True

        # ----------------- CIRCLE FIT -----------------
        if remaining_points.shape[0] >= min_circle_pts:
            # Pick 3 points spread across indices
            idx = np.linspace(0, remaining_points.shape[0]-1, 3, dtype=int)
            sample_pts = remaining_points[idx]
            center, radius = fit_circle(sample_pts)

            if circle_r_range[0] <= radius <= circle_r_range[1]:
                distances_circle = np.abs(point_circle_distance(center, radius, remaining_points))
                circle_inliers = remaining_points[distances_circle <= distance_thresh_circle]

                if circle_inliers.shape[0] >= min_circle_pts:
                    # Re-estimate circle from all inliers
                    center, radius = fit_circle(circle_inliers)
                    circles.append((center, radius, circle_inliers))
                    remaining_points = remaining_points[distances_circle > distance_thresh_circle]
                    new_segment_found = True

        if not new_segment_found:
            break

    return lines, circles, remaining_points
```

```python
def plot_residual_histogram(points, model_type="line", model=None):
    if model_type=="line":
        residuals = np.array([point_line_distance(p, model) for p in points])
    else:
        residuals = np.array([point_circle_distance(p, model) for p in points])
    plt.hist(residuals, bins=50)
    plt.xlabel("Distance (m)")
    plt.ylabel("Frequency")
    plt.title(f"{model_type} residual histogram")
    plt.show()
```

```python
def save_ply(filename, points, lines, circles):
    with open(filename, 'w') as f:
        n_points = len(points)
        f.write(f"ply\nformat ascii 1.0\nelement vertex {n_points}\n")
        f.write("property float x\nproperty float y\nproperty float z\n")
        f.write("property uchar red\nproperty uchar green\nproperty uchar blue\nend_header\n")
        for p in points:
            color = [255,255,255]  # default white
            for _, inliers in lines:
                if any(np.all(p==pt) for pt in inliers):
                    color = [0,0,255]  # blue line
            for _, inliers in circles:
                if any(np.all(p==pt) for pt in inliers):
                    color = [255,0,0]  # red circle
            f.write(f"{p[0]} {p[1]} 0 {color[0]} {color[1]} {color[2]}\n")

save_ply("lidar_result.ply", points, lines, circles)
```
